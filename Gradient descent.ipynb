```python
class GD:
    def __init__(self,coef):
        self.coef = coef
    def cost_derivative(self,y,ypred):
        return ypred-y
    

    def gradient_descent(self,x,y,epochs,lr=0.00000001):
        for _ in range(epochs):
            ypred=np.dot(x,self.coef)
            cost_prime = self.cost_derivative(y,ypred)
            step_size = lr *(1/len(y))* np.dot(x.T,cost_prime)

            self.coef = self.coef - step_size
            


            cost_function = np.sqrt(0.5 * np.mean(np.square(np.dot(x,self.coef)-y)))
            print(cost_function)

        return self.coef               
  
                       
```
